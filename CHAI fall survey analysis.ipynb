```python
df = df[df['Q4'] != 's']
```

```python
df.rename(columns={'Q3': 'Education', 'Q116':'Sector',
                   'Q119':'Non-profit', 'Q117':'Role in AI',
                   'Q4':'Email', 'Q77':'Expertise'
                   }, inplace=True)
```

```python
df['id'] = df['Email'].rank(method='dense', ascending=False, na_option='bottom').astype(int)
```

```python
df.head()
```

```python
df_cleaned = df.drop_duplicates(subset='Email')
```

```python
df_cleaned.info()
```

```python
duplicates_count = df_cleaned['Email'].value_counts()
duplicates_count
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)
selected_variables = ['Email','Education', 'Sector', 'Non-profit', 'Role in AI', 'id']
intro_data = df_cleaned[selected_variables]
```

```python
intro_data.info()
# intro_data.head(40)
```

```python
intro_data.head()
```

```python
intro_data['Education'] = intro_data['Education'].astype(str)

# create dummy variables
dummy_df = intro_data['Education'].str.get_dummies(',')

intro_data = pd.concat([intro_data, dummy_df], axis=1)
```

```python
intro_data.rename(columns={'1': 'High-school', '2':'Bachelor',
                   '3':'Master', '4':'PhD',
                   '5':'JD', '6':'PharmD', '7':'MD'
                   }, inplace=True)
# intro_data = intro_data.drop(['nan'], axis=1)
intro_data.head()
```

```python
# pivot_table to reshape the DataFrame
pivot_df = df.pivot_table(index='Email', columns=df.groupby('Email').cumcount(), values='Expertise', aggfunc='first')

pivot_df.columns = [f'Expertise_{i+1}' for i in range(pivot_df.shape[1])]

pivot_df.reset_index(inplace=True)
result_df = pd.merge(pivot_df, df[['Email']].drop_duplicates(), on='Email')

print(result_df)
```

```python
intro_data = pd.merge(result_df, intro_data, on='Email')
```

```python
def count_answers(row):
    count = sum(~pd.isnull(row))
    return 2 if count == 2 else 1

# Apply the function to create the 'Number of answers' column
intro_data['Number of answers'] = intro_data[['Expertise_1', 'Expertise_2']].apply(count_answers, axis=1)
```

```python
# intro_data = intro_data[intro_data['Email'] != 's']
intro_data.head()
```

```python
def update_expertise(row):
    if pd.isnull(row['Expertise_1']) and not pd.isnull(row['Expertise_2']):
        row['Expertise_1'] = row['Expertise_2']
        row['Expertise_2'] = np.nan
    return row

# Apply the function to update the values
intro_data = intro_data.apply(update_expertise, axis=1)
```
