### Usefulness

```python
stage1_usefulness = filtered_stage1.filter(regex='^(stage1_usefulness)', axis=1)
```

```python
stage1_usefulness.head()
```

```python
stage1_usefulness.dropna(subset=['stage1_usefulness_1'], inplace=True)

stage1_usefulness.rename(columns={
    'stage1_usefulness_1': 'Clearly define the problem to be solved<br>and why the AI tool is necessary',
    'stage1_usefulness_2': 'Consider the magnitude and frequency of<br>benefits and risks',
    'stage1_usefulness_3': 'Consider if clinical end users are likely to put<br>trust in the model and its output',
    'stage1_usefulness_4': 'The user (or applicable users) must be identified',
    'stage1_usefulness_5': 'Consider the AI-specific Cost-Benefit Analysis (CBA)',
    'stage1_usefulness_6': 'Consider the documentation burden created<br>from the implementation of AI',
    'stage1_usefulness_7': 'Consider how the AI technology will<br>integrate into clinical workflow',
    'stage1_usefulness_8': 'Consider system capacities and how the<br>AI may affect those',
    'stage1_usefulness_9': 'Review previous effectiveness and evidence<br>of success elsewhere',
    'stage1_usefulness_10': 'Consider the information needs of the user',
    'stage1_usefulness_11': 'Consider how the AI will integrate with<br>existing technologies',
    'stage1_usefulness_12': 'Consider how end users will be engaged in<br>the design of the algorithm',
    'stage1_usefulness_13': 'Consider the organizational capabilities<br>required to effectively use and manage the AI'
}, inplace=True)

stage1_usefulness.reset_index(drop=True)
# stage1_usefulness.head(50)
```

```python
stage1_usefulness_m = pd.melt(stage1_usefulness, value_name='index', var_name='Consideration')
stage1_usefulness_m
stage1_usefulness_r = pd.crosstab(stage1_usefulness_m['Consideration'], stage1_usefulness_m['index'])


stage1_usefulness_r.reset_index(inplace=True)
stage1_usefulness_r.head()
```

```python
stage1_usefulness_r.rename(columns={1: 'Not at all important',
                   2 : 'Slightly important',
                   3 : 'Moderately important',
                   4 : 'Very important',
                   5 : 'Extremely important',
                   }, inplace=True)
```

```python
import plotly.offline as pyo
import plotly.graph_objects as go
import plotly.io as pio


stage1_usefulness_sort = stage1_usefulness_r.sort_values(by='Extremely important', ascending = True)
category_order = [
    'Not at all important',
'Slightly important',
'Moderately important',
'Very important',
'Extremely important',
]

considerations = stage1_usefulness_sort['Consideration'].tolist()
stage1_usefulness_sort = stage1_usefulness_sort[category_order]
# stage1_usefulness_sort

fig = go.Figure()
for column in stage1_usefulness_sort.columns:
    fig.add_trace(go.Bar(
        x = stage1_usefulness_sort[column],
        y = considerations,
        name = column,
        orientation = 'h',
        marker_color = cl.scales[str(len(category_order))]['div']['RdYlGn'][category_order.index(column)],
         hovertemplate = '<b>Answers</b>: %{x}')
    )
    
# Calculate the position for the vertical line at 35%
total_observations = len(stage1_usefulness)
observation_percentage = 0.35
vertical_line_position_35 = observation_percentage * total_observations
fig.add_vline(x=vertical_line_position_35, 
              line_width=2, 
              line_dash="dash", 
              line_color="red", 
            annotation_text="35%", 
              annotation_position="top left",
              annotation_font_size=16,
              annotation_font_color="red"              
              )

observation_percentage = 0.5
vertical_line_position_50 = observation_percentage * total_observations
fig.add_vline(x=vertical_line_position_50,               
              line_width=2, 
              line_dash="dash", 
              line_color="blue", 
            annotation_text="50%", 
              annotation_position="top right",
              annotation_font_size=16,
              annotation_font_color="blue"
              )
titlevar=(f'Stage 1 - Usefulness (N={total_observations})')
    

fig.update_layout(
    autosize=False,
    width=1050,
    height=775,
    yaxis = dict(
tickfont = dict(size=11.5)),    
    barmode = 'relative',
    title = titlevar,
    xaxis_title='Number of answers',
    yaxis_title='Consideration',
    hovermode="closest",
    hoverlabel=dict(
namelength = -1,
        bgcolor="white",
        font_size=14,
        font_family="Arial"
    )
)



# Display the figure and save it to an HTML file
pyo.plot(fig, filename='../Outputs/Charts/stage1_usefulness_stackedbar_dataScience.html')
```
